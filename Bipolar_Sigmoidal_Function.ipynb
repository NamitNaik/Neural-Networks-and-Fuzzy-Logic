```
import numpy as np

import matplotlib.pyplot as plt

def bip_sigmoid(x,lambda1):
  y=(1-np.exp(-lambda1*x))/(1+np.exp(-lambda1*x))
  return y

x=np.linspace(-20,20,200)

for lambda1 in range(1,5):
  y1=bip_sigmoid(x,lambda1)
  plt.plot(x,y1)

plt.xlabel("net input")
plt.ylabel("output")
plt.title("Bipolar sigmoidal function")
plt.text(20,1,r"Bipolar Sigmoidal Function: $f(x)=\frac{1-e^-lambda*x}{1+e^-lambda*x}$")
plt.grid()

```
