```
import numpy as np

import matplotlib.pyplot as plt

def bipstep_fn(i):
  if i>=0:
    return 1
  else:
    return -1        
  
    
x=np.linspace(-50,50,200)
y1=list()

for i in x:
    y=bipstep_fn(i)
    y1.append(y)


plt.plot(x,y1,'r')
plt.xlabel("Net Input")
plt.ylabel("Output")
plt.title("Bipolar Step Function")
plt.text(40,1,r"Bipolar Step Function: $f(x)=1, for x>=0, f(x)=-1, for x<0$")
plt.grid()
```
