```
import numpy as np

import matplotlib.pyplot as plt

def identity_fn(x1):
  y=x1
  return y

x1=np.linspace(0.1,20,50)
y1=identity_fn(x1)

plt.plot(x1,y1,'r')
plt.xlabel("Net Input")
plt.ylabel("Output")
plt.title("Identity Function")
plt.text(20,20,r"Identity Function: $f(x)=x$")
plt.grid()
```
