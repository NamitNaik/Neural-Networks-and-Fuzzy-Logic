```
import numpy as np

import matplotlib.pyplot as plt

def ramp_fn(i):
  if i>1:
    return 1
  if i>=0 and i<=1:
    return i
  else:
    return 0            
  
    
x=np.linspace(-2,2,1000)
y1=list()

for i in x:
    y=ramp_fn(i)
    y1.append(y)


plt.plot(x,y1,'r')
plt.xlabel("Net Input")
plt.ylabel("Output")
plt.title("Ramp Function")
plt.text(2,1,r"Ramp Function: $f(x)=0, for x < 0, f(x)=x for 0<=x<=1, f(x)=1 for x>1$")
plt.grid()
```
