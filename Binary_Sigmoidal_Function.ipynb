```
import numpy as np

import matplotlib.pyplot as plt

def bin_sigmoid(x,lambda1):
  y=1/(1+np.exp(-lambda1*x))
  return y

x=np.linspace(-20,20,100)

for lambda1 in range(1,5):
  y1=bin_sigmoid(x,lambda1)
  plt.plot(x,y1)
  

plt.xlabel("Net Input")
plt.ylabel("Output")
plt.title("Binary Sigmoidal Function")
plt.text(20,1,r"Binary Sigmoidal Function: $f(x)=\frac{1}{1+e^-lambda*x}$")
plt.grid()
```
